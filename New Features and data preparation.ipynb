```python
from collections import defaultdict
import pandas as pd

file_path = "labels.csv"
columns_to_drop = ["storyId","word_index"]
labels = pd.read_csv(file_path, usecols=lambda column: column not in columns_to_drop)

pd.set_option('display.max_rows', None)


label_dict = defaultdict(lambda: defaultdict(list))
for index, row in labels.iterrows():
    activity = row["activityId"]
    label = row["label"]
    phrase_index = row["phraseIndex"]
    label_dict[activity][phrase_index].append(label)
print(label_dict)
```

```python
import json

def load_arpabet_to_amirabet(filepath):
    with open(filepath, 'r') as f:
        return json.load(f)

def load_all_story_words(filepath):
    with open(filepath, 'r') as f:
        return {line.split()[0]: ' '.join(line.split()[1:]) for line in f}

def get_phonetic_expression(sentence, arpabet_to_amirabet, all_story_words):
    words = sentence.split()
    phonetic_expression = []
    for word in words:
        if word.upper() in all_story_words:
            arpabet = all_story_words[word.upper()]
            amirabet = [arpabet_to_amirabet[arp] for arp in arpabet.split() if arp in arpabet_to_amirabet]
            phonetic_expression.append(''.join(amirabet))
    return ' '.join(phonetic_expression)

def get_sentence_phonetic_expression(sentence):
    arpabet_to_amirabet_filepath = 'arpabet_to_amirabet.json'
    all_story_words_filepath = 'all_story_words.dic'

    arpabet_to_amirabet = load_arpabet_to_amirabet(arpabet_to_amirabet_filepath)
    all_story_words = load_all_story_words(all_story_words_filepath)

    return get_phonetic_expression(sentence, arpabet_to_amirabet, all_story_words)

```

```python
# Example usage
file_path = "asr_data.csv"  
asr = pd.read_csv(file_path)
print(len(asr))

expected_phoneme_dict = defaultdict(lambda: defaultdict(list))
for index, row in asr.iterrows():
    story_text = row["story_text"]
    activity = row["activityId"]
    phrase_index = row["phrase_index"]
    expression = get_sentence_phonetic_expression(row["story_text"])
    print(expression)
    expected_phoneme_dict[activity][phrase_index].append(expression)
    
```

```python
file_path = "asr_data.csv"  
asr = pd.read_csv(file_path)
print(len(asr))

import json
num_rows_without_conf = 0
num_rows_with_conf = 0
desired_length = 50

# New DataFrame with specified columns
new_df = pd.DataFrame(columns=[
    'activityId',
    'phrase_index',
    'expected_phonemes',
    'actual_phonemes',
    'conf_array',
    'time_taken_list',
    'actual_labels_list',
    'is_length_of_spoken_and_expected_words_same'
])
for index, row in asr.iterrows():
    activity = row["activityId"]
    kaldina_data = row['kaldiNa_data']
    story_text = row['story_text']
    wave2vec_phonemes = row['wav2vec_transcript_phonemes']
    phrase_index = row["phrase_index"]
    if pd.isna(row["kaldiNa_data"]):
        num_rows_without_conf += 1
        continue
        
    try:
        kaldiNa_data_json = json.loads(row["kaldiNa_data"])
        
        if 'raw' in kaldiNa_data_json and 'ref-conf' in kaldiNa_data_json['raw'] and pd.notna(wave2vec_phonemes):
            conf = kaldiNa_data_json['raw']['ref-conf']
            conf_array = [entry[1] for entry in conf]
            time_taken_list = []
            
            try:
                kaldi_data_json = eval(kaldina_data)  # Convert JSON string to dictionary
                for word_data in kaldi_data_json['transcription']:
                    time_taken = word_data['end_time'] - word_data['start_time']
                    time_taken_list.append(time_taken)
                if len(wave2vec_phonemes.split()) == len(story_text.split()):
                    is_length_of_spoken_and_expected_words = 1
                else:
                    is_length_of_spoken_and_expected_words = 0
                labels_list = label_dict[activity][phrase_index]
                expected_phonemes = expected_phoneme_dict[activity][phrase_index][0]
                conf_array += [-1] * (desired_length - len(conf_array))
                time_taken_list += [-1] * (desired_length - len(time_taken_list))
                labels_list += [-1] * (desired_length - len(labels_list))
                new_row = {
                    'activityId': row['activityId'],
                    'phrase_index': row['phrase_index'],
                    'expected_phonemes': expected_phonemes,
                    'actual_phonemes': wave2vec_phonemes,
                    'conf_array': conf_array,  
                    'time_taken_list': time_taken_list,  
                    'actual_labels_list': labels_list,  
                    'is_length_of_spoken_and_expected_words_same': is_length_of_spoken_and_expected_words
                }
                new_df.loc[len(new_df)] = new_row
                num_rows_with_conf += 1
            except:
                print(len(kaldina_data))
                num_rows_without_conf += 1
            
        else:
            num_rows_without_conf += 1
    except KeyError:
        num_rows_without_conf += 1
    


```
