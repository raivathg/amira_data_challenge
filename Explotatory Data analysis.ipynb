```python
import pandas as pd
import matplotlib.pyplot as plt

labels_df = pd.read_csv("labels.csv")

label_counts = labels_df['label'].value_counts()

# Create a bar plot
plt.bar(label_counts.index, label_counts.values)

# Add labels and title
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Distribution of Labels')

# Add label annotations
for i, count in enumerate(label_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

# Add legend
plt.legend(['Incorrect (0)', 'Correct (1)'])

# Show plot
plt.show()
```

```python
import json

def load_arpabet_to_amirabet(filepath):
    with open(filepath, 'r') as f:
        return json.load(f)

def load_all_story_words(filepath):
    with open(filepath, 'r') as f:
        return {line.split()[0]: ' '.join(line.split()[1:]) for line in f}

def get_phonetic_expression(sentence, arpabet_to_amirabet, all_story_words):
    words = sentence.split()
    phonetic_expression = []
    for word in words:
        if word.upper() in all_story_words:
            arpabet = all_story_words[word.upper()]
            amirabet = [arpabet_to_amirabet[arp] for arp in arpabet.split() if arp in arpabet_to_amirabet]
            phonetic_expression.append(''.join(amirabet))
    return ' '.join(phonetic_expression)

arpabet_to_amirabet_filepath = 'arpabet_to_amirabet.json'
all_story_words_filepath = 'all_story_words.dic'

arpabet_to_amirabet = load_arpabet_to_amirabet(arpabet_to_amirabet_filepath)
all_story_words = load_all_story_words(all_story_words_filepath)

sentence = "I love water. I enjoy playing with water"

phonetic_expression = get_phonetic_expression(sentence, arpabet_to_amirabet, all_story_words)
print(phonetic_expression)
```

```python

import json

def load_amirabet_to_arpabet(filepath):
    with open(filepath, 'r') as f:
        return json.load(f)

def load_all_story_words(filepath):
    with open(filepath, 'r') as f:
        return {line.split()[0]: ' '.join(line.split()[1:]) for line in f}

def convert_phonetic_to_words(phonetic_expression, inverted_amirabet_to_arpabet, all_story_words):
    arpabet_to_word = {tuple(arp.split()): word for word, arp in all_story_words.items()}
    words = []
    for phonetic_word in phonetic_expression.split():
        phonetic_symbols = phonetic_word.split()
        arpabet_symbols = [inverted_amirabet_to_arpabet.get(symbol, '') for symbol in phonetic_symbols[0]]
        arpabet_phonemes = ' '.join(arpabet_symbols)
        print(arpabet_phonemes)
        if arpabet_phonemes:
            word = arpabet_to_word.get(tuple(arpabet_phonemes.split()), None)
            if word:
                words.append(word)
    return ' '.join(words)


amirabet_to_arpabet_filepath = 'arpabet_to_amirabet.json'
all_story_words_filepath = 'all_story_words.dic'

amirabet_to_arpabet = load_amirabet_to_arpabet(amirabet_to_arpabet_filepath)
inverted_amirabet_to_arpabet = {value: key for key, value in amirabet_to_arpabet.items()}
all_story_words = load_all_story_words(all_story_words_filepath)

phonetic_expression = "γ lʌv wɔtɝw laz γ ɛnjω plaɪŋ wɪθ"

words = convert_phonetic_to_words(phonetic_expression, inverted_amirabet_to_arpabet, all_story_words)
print(words)
```
